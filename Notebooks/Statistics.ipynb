# Pitfalls

## Area Bias
Area bias, also known as geographical bias, occurs when the sample under consideration is not representative of the entire population because it is limited to a specific geographic area. This can lead to inaccurate generalizations about the broader population.

## Self-Selection Bias
Self-selection bias occurs when individuals voluntarily choose to participate in a study, leading to a non-random sample. This can result in skewed or unrepresentative data, as certain groups may be overrepresented or underrepresented.

## Leading Question Bias
Leading question bias occurs when the wording of a question suggests a particular answer or influences the respondent's understanding. This can lead to inaccurate or misleading responses.

## Social Desirability Bias
Social desirability bias occurs when respondents provide answers that they believe are socially acceptable or desirable, rather than expressing their true opinions or behaviors. This bias often stems from a desire to conform to societal norms or present oneself in a favorable light.

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulating a six-sided die roll
num_trials = 1000
dice_rolls = np.random.randint(1, 7, size=num_trials)

# Calculating the probability distribution
unique_outcomes, counts = np.unique(dice_rolls, return_counts=True)
probabilities = counts / num_trials

# Creating a bar plot
plt.bar(unique_outcomes, probabilities, color='skyblue')
plt.xlabel('Outcome on the Die')
plt.ylabel('Probability')
plt.title('Probability Distribution of a Six-Sided Die')
plt.xticks(np.arange(1, 7))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, poisson

# Function to calculate variance
def calculate_variance(data):
    mean = np.mean(data)
    return np.mean((data - mean)**2)

# Function to calculate standard deviation
def calculate_std_dev(data):
    return np.sqrt(calculate_variance(data))

# Function to calculate coefficient of variation
def calculate_coeff_of_var(data):
    mean = np.mean(data)
    std_dev = calculate_std_dev(data)
    return (std_dev / mean) * 100

# Generate a sample dataset from a normal distribution
np.random.seed()
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Generate a sample dataset from a Poisson distribution
np.random.seed()
lambda_poisson = 3
data_poisson = np.random.poisson(lambda_poisson, size=1000)

# Calculate statistics for normal distribution
variance_normal = calculate_variance(data_normal)
std_deviation_normal = calculate_std_dev(data_normal)
q1_normal, q3_normal = np.percentile(data_normal, [25, 75])
iqr_value_normal = q3_normal - q1_normal
coeff_of_var_normal = calculate_coeff_of_var(data_normal)

# Calculate statistics for Poisson distribution
variance_poisson = calculate_variance(data_poisson)
std_deviation_poisson = calculate_std_dev(data_poisson)
q1_poisson, q3_poisson = np.percentile(data_poisson, [25, 75])
iqr_value_poisson = q3_poisson - q1_poisson
coeff_of_var_poisson = calculate_coeff_of_var(data_poisson)

# Create subplots with 2 rows and 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 12), gridspec_kw={'width_ratios': [2, 1]})

# Normal distribution plot
axes[0, 0].hist(data_normal, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
xmin, xmax = axes[0, 0].get_xlim()
x_normal = np.linspace(xmin, xmax, 100)
p_normal = norm.pdf(x_normal, np.mean(data_normal), np.std(data_normal))
axes[0, 0].plot(x_normal, p_normal, 'k', linewidth=2)
axes[0, 0].axvline(np.mean(data_normal), color='red', linestyle='dashed', linewidth=2, label='Mean')
axes[0, 0].axvline(np.mean(data_normal) - std_deviation_normal, color='orange', linestyle='dashed', linewidth=2, label='1 Std Dev Below Mean')
axes[0, 0].axvline(np.mean(data_normal) + std_deviation_normal, color='orange', linestyle='dashed', linewidth=2, label='1 Std Dev Above Mean')
axes[0, 0].axvline(q1_normal, color='green', linestyle='dashed', linewidth=2, label='Q1')
axes[0, 0].axvline(q3_normal, color='green', linestyle='dashed', linewidth=2, label='Q3')
axes[0, 0].axvline(np.mean(data_normal) - iqr_value_normal/2, color='purple', linestyle='dashed', linewidth=2, label='Median - IQR/2')
axes[0, 0].axvline(np.mean(data_normal) + iqr_value_normal/2, color='purple', linestyle='dashed', linewidth=2, label='Median + IQR/2')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Probability Density')
axes[0, 0].set_title('Normal Distribution with Statistics')
axes[0, 0].legend()

# Boxplot for normal distribution
axes[0, 1].boxplot(data_normal, vert=False)
axes[0, 1].set_yticklabels([])
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_title('Boxplot - Normal Distribution')

# Poisson distribution plot
axes[1, 0].hist(data_poisson, bins=np.arange(0, max(data_poisson) + 1.5) - 0.5, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
x_poisson = np.arange(0, max(data_poisson) + 1)
p_poisson = poisson.pmf(x_poisson, lambda_poisson)
axes[1, 0].plot(x_poisson, p_poisson, 'k', linewidth=2)
axes[1, 0].axvline(np.mean(data_poisson), color='red', linestyle='dashed', linewidth=2, label='Mean')
axes[1, 0].axvline(np.mean(data_poisson) - std_deviation_poisson, color='orange', linestyle='dashed', linewidth=2, label='1 Std Dev Below Mean')
axes[1, 0].axvline(np.mean(data_poisson) + std_deviation_poisson, color='orange', linestyle='dashed', linewidth=2, label='1 Std Dev Above Mean')
axes[1, 0].axvline(q1_poisson, color='green', linestyle='dashed', linewidth=2, label='Q1')
axes[1, 0].axvline(q3_poisson, color='green', linestyle='dashed', linewidth=2, label='Q3')
axes[1, 0].axvline(np.mean(data_poisson) - iqr_value_poisson/2, color='purple', linestyle='dashed', linewidth=2, label='Median - IQR/2')
axes[1, 0].axvline(np.mean(data_poisson) + iqr_value_poisson/2, color='purple', linestyle='dashed', linewidth=2, label='Median + IQR/2')
axes[1, 0].set_xlabel('Value')
axes[1, 0].set_ylabel('Probability Density')
axes[1, 0].set_title('Poisson Distribution with Statistics')
axes[1, 0].legend()

# Boxplot for Poisson distribution
axes[1, 1].boxplot(data_poisson, vert=False)
axes[1, 1].set_yticklabels([])
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_title('Boxplot - Poisson Distribution')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson, binom, geom, randint

# Set a random seed for reproducibility
np.random.seed(2)

# Generate data from a Poisson distribution
lambda_poisson = 3
data_poisson = np.random.poisson(lambda_poisson, size=1000)

# Generate data from a Binomial distribution
n_binom, p_binom = 10, 0.3
data_binom = np.random.binomial(n_binom, p_binom, size=1000)

# Generate data from a Geometric distribution
p_geom = 0.2
data_geom = np.random.geometric(p_geom, size=1000)

# Generate data from a Discrete Uniform distribution
low_uniform, high_uniform = 1, 6
data_uniform = np.random.randint(low_uniform, high_uniform + 1, size=1000)

# Plot histograms for all distributions
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Poisson distribution
axes[0, 0].hist(data_poisson, bins=np.arange(0, max(data_poisson) + 1.5) - 0.5, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
x_poisson = np.arange(0, max(data_poisson) + 1)
p_poisson = poisson.pmf(x_poisson, lambda_poisson)
axes[0, 0].vlines(x_poisson, 0, p_poisson, colors='k', linewidth=2)
axes[0, 0].set_title('Poisson Distribution')

# Binomial distribution
axes[0, 1].hist(data_binom, bins=np.arange(0, n_binom + 1.5) - 0.5, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
x_binom = np.arange(0, n_binom + 1)
p_binom = binom.pmf(x_binom, n_binom, p_binom)
axes[0, 1].vlines(x_binom, 0, p_binom, colors='k', linewidth=2)
axes[0, 1].set_title('Binomial Distribution')

# Geometric distribution
axes[1, 0].hist(data_geom, bins=np.arange(0, max(data_geom) + 1.5) - 0.5, density=True, alpha=0.7, color='skyblue', edgecolor='black')
x_geom = np.arange(0, max(data_geom) + 1)
p_geom = geom.pmf(x_geom, p_geom)
axes[1, 0].vlines(x_geom, 0, p_geom, colors='k', linewidth=2)
axes[1, 0].set_title('Geometric Distribution')

# Discrete Uniform distribution
axes[1, 1].hist(data_uniform, bins=np.arange(low_uniform - 0.5, high_uniform + 1.5) - 0.5, density=True, alpha=0.7, color='lightyellow', edgecolor='black')
x_uniform = np.arange(low_uniform, high_uniform + 1)
p_uniform = np.full_like(x_uniform, 1/(high_uniform - low_uniform + 1))
axes[1, 1].vlines(x_uniform, 0, p_uniform, colors='k', linewidth=2)
axes[1, 1].set_title('Discrete Uniform Distribution')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
from ipywidgets import interact, widgets

# Function to plot the binomial distribution
def plot_binomial_distribution(probability):
    n = 20  # Number of trials
    x = np.arange(0, n + 1)
    
    # Calculate binomial PMF
    pmf = binom.pmf(x, n, probability)
    
    #Check if skewed, depends on probability
    title = 'symmetric distribution'
    if probability > 0.5: 
        title = "negatively skewed, larger left tale"
    elif probability < 0.5:
        title = "positively skewed, larger right tale"
        
    # Plot the distribution
    plt.bar(x, pmf, color='lightblue', edgecolor='black')
    plt.suptitle(f'Binomial Distribution (n={n}, p={probability:.2f})')
    plt.title(f'{title}')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.xticks(x)
    plt.show()

# Create an interactive slider for probability
probability_slider = widgets.FloatSlider(value=0.5, min=0, max=1, step=0.01, description='Probability (p)')

# Create an interactive plot
interact(plot_binomial_distribution, probability=probability_slider)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson
from ipywidgets import interact, widgets

# Function to plot the Poisson distribution
def plot_poisson_distribution(lambda_param):
    x = np.arange(0, 20)
    
    # Calculate Poisson PMF
    pmf = poisson.pmf(x, lambda_param)
    
    # Plot the distribution
    plt.bar(x, pmf, color='lightcoral', edgecolor='black')
    plt.title(f'Poisson Distribution (={lambda_param:.2f})')
    plt.xlabel('Number of Events')
    plt.ylabel('Probability')
    plt.xticks(x)
    plt.show()

# Create an interactive slider for lambda
lambda_slider = widgets.FloatSlider(value=2, min=0.1, max=10, step=0.1, description='Lambda ()')

# Create an interactive plot
interact(plot_poisson_distribution, lambda_param=lambda_slider)
```

# Hypergeometric <a id='Hypergeometirc'></a>

A crate contains 100 light bulbs of  which 15 are defective and 85 are not. A Quality Control Inspector randomly samples 10 bulbs without replacement. If all bulbs are working, the lot is accepted. What are the chances of this happening?

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import hypergeom

# Parameters
total_bulbs = 100
defective_bulbs = 15
sample_size = 10

# Calculate the probability of all bulbs being working
non_defective_bulbs = total_bulbs - defective_bulbs
probability = hypergeom.pmf(0, total_bulbs, non_defective_bulbs, sample_size)

# Plot the probability mass function (PMF)
x = np.arange(0, sample_size + 1)
pmf_values = hypergeom.pmf(x, total_bulbs, non_defective_bulbs, sample_size)

plt.bar(x, pmf_values, color='lightgreen', edgecolor='black')
plt.title(f'Probability of working bulbs in a sample of {sample_size}')
plt.xlabel('Number of working bulbs')
plt.ylabel('Probability')

# Add labels above the bars
for i, prob in enumerate(pmf_values):
    plt.text(i, prob, f'{prob:.4f}', ha='center', va='bottom', fontsize=7)

plt.show()

print(f"The probability of all bulbs being working is: {probability:.4f}")

```
