# Introduction to Statistics in Python

Statistics is a powerful tool for analyzing and interpreting data. In this Jupyter Notebook, we will explore some fundamental statistical concepts using Python. Python provides various libraries that make statistical analysis efficient and accessible.

## Libraries Used

We will primarily use the following Python libraries for our statistical analysis:

1. **NumPy:** NumPy is the fundamental package for scientific computing with Python. It provides support for large, multi-dimensional arrays and matrices, along with mathematical functions to operate on these arrays.

2. **Pandas:** Pandas is a fast, powerful, and flexible open-source data analysis and manipulation tool. It provides data structures like DataFrames that are ideal for working with structured data.

3. **Matplotlib and Seaborn:** These libraries are used for data visualization. Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations, while Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics.

## Topics Covered

We will cover the following topics in this notebook:

1. **Descriptive Statistics:** Calculating measures of central tendency (mean, median, mode) and measures of dispersion (range, variance, standard deviation).

2. **Data Distributions:** Understanding different types of distributions, such as normal distribution, and visualizing them.

3. **Hypothesis Testing:** Performing hypothesis tests to make inferences about populations based on sample data.

4. **Correlation and Regression:** Analyzing relationships between variables using correlation and regression analysis.

5. **Statistical Plots:** Creating various statistical plots to visualize data distributions and relationships.

Throughout the notebook, we will provide code examples and explanations to help you understand and apply these statistical concepts using Python.


```python
# Descriptive Statistics
import numpy as np

# Sample data
data = np.array([10, 15, 20, 25, 30, 35, 40])

# Mean, Median, Mode
mean_value = np.mean(data)
median_value = np.median(data)
mode_value = np.argmax(np.bincount(data))

print(f"Mean: {mean_value}, Median: {median_value}, Mode: {mode_value}")
```

```python
# Data Distributions
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a random normal distribution
data_normal = np.random.normal(size=1000)

# Plot histogram
plt.hist(data_normal, bins=30, color='skyblue', edgecolor='black')
plt.title('Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Hypothesis Testing
from scipy.stats import ttest_ind

# Sample data for two groups
group1 = np.array([25, 30, 35, 40, 45])
group2 = np.array([20, 22, 25, 28, 30])

# Independent t-test
t_stat, p_value = ttest_ind(group1, group2)

print(f"T-statistic: {t_stat}, P-value: {p_value}")


'''
The t-statistic measures the difference between the means of the two groups, 
and the p-value indicates the probability of observing such a difference by random chance
'''
```

```python
# Correlation and Regression
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sns

# Sample data
data = {'X': [1, 2, 3, 4, 5], 'Y': [2, 4, 5, 4, 5]}
df = pd.DataFrame(data)

# Pearson correlation
correlation, _ = pearsonr(df['X'], df['Y'])
print(f"Pearson Correlation: {correlation}")

# Regression plot
sns.regplot(x='X', y='Y', data=df)
plt.title('Regression Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Gegeven parameters
mu_kosten = 850000
sigma_kosten = 170000
opbrengst = 1000000

# a. Kans dat het project winstgevend is
prob_winstgevend = norm.cdf(opbrengst, loc=mu_kosten, scale=sigma_kosten)

# b. Kans dat het project verlies oplevert
prob_verlies = 1 - prob_winstgevend

# Bereken de Z-score
z_score = (opbrengst - mu_kosten) / sigma_kosten

# Genereer gegevens voor het histogram
x = np.linspace(mu_kosten - 4 * sigma_kosten, mu_kosten + 4 * sigma_kosten, 1000)
y = norm.pdf(x, loc=mu_kosten, scale=sigma_kosten)

# Maak een histogram en voeg de Z-score toe
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normale verdeling')
plt.fill_between(x, y, where=(x <= opbrengst), color='orange', alpha=0.5, label='Kans op winstgevendheid')
plt.axvline(x=opbrengst, color='red', linestyle='--', label=f'Opbrengst (Z-score = {z_score:.2f})')
plt.title('Kansverdeling van projectkosten')
plt.xlabel('Kosten (Euro)')
plt.ylabel('Kansdichtheid')
plt.legend()
plt.grid(True)
plt.show()

print(f"a. Kans dat het project winstgevend is: {prob_winstgevend:.4f}")
print(f"b. Kans dat het project verlies oplevert: {prob_verlies:.4f}")
print(f"Z-score: {z_score:.2f}")
```

# Oefening 3.

Als een orthodontist een plaatjesbeugel aanbrengt, moet hij op het tandoppervlak een plek droog houden. Nu is er een nieuwe lijm ontwikkeld die ook kan worden aangebracht op een niet-droge plek.  Men vreest echter dat de nieuwe lijm niet zo sterk is als de tot nu toe gebruikelijke lijm.  Bij een test op 10 (getrokken) tanden bleek de gemiddelde treksterkte na 24 uur gelijk te zijn aan 5,07 MPa met een standaardafwijking van 0,46 MPa. Orthodontisten willen weten of de nieuwe lijm minder sterk is dan 5,70 MPa, de verwachte treksterkte van de tot nu toe gebruikelijke lijm.
1. Formuleer de nulhypothese en de alternatieve hypothese voor de toe te passen statistische toets.
1. Bepaal het kritieke gebied bij een significantieniveau van  = 0,01.
1. Bereken de uitkomst van de toetsingsgrootheid.
1. Als we aannemen dat aan de voorwaarden is voldaan, wat is dan je conclusie op grond van deze toets?

```python
from scipy.stats import t

# Gegeven waarden
steekproefgemiddelde = 5.07
verwachte_gemiddelde = 5.70
standaardafwijking = 0.46
steekproefgrootte = 10
significantieniveau = 0.01

# Standaardfout berekenen
standaardfout = standaardafwijking / (steekproefgrootte ** 0.5)

# Toetsingsgrootheid berekenen (t-waarde)
t_waarde = (steekproefgemiddelde - verwachte_gemiddelde) / standaardfout

# Kritieke grens berekenen
kritieke_grens = t.ppf(significantieniveau, df=steekproefgrootte - 1)

# Beslissing nemen op basis van de t-waarde
conclusie = "Verwerp de nulhypothese" if t_waarde < kritieke_grens else "Accepteer de nulhypothese"

print(f"t-waarde: {t_waarde:.4f}")
print(f"Kritieke grens: {kritieke_grens:.4f}")
print(f"Conclusie: {conclusie}")

# Kansen voor de t-verdeling
x_t = np.linspace(-3, 3, 1000)
y_t = t.pdf(x_t, df=steekproefgrootte - 1)

# t-waarde markeren
plt.plot(x_t, y_t, label=f't-verdeling (df={steekproefgrootte-1})')
plt.fill_between(x_t, y_t, where=(x_t <= kritieke_grens), color='red', alpha=0.3, label='Verwerpingsgebied')

# Kritieke grens markeren
plt.axvline(x=kritieke_grens, color='black', linestyle='--', label='Kritieke grens')

plt.title('t-toets voor Treksterkte met Kritieke Grens')
plt.xlabel('t-waarde')
plt.ylabel('Kansdichtheid')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
sample_mean = -75.4
sample_std_dev = 10.9
sample_size = 25
confidence_level = 0.90

# Calculate the standard error of the mean
standard_error = sample_std_dev / np.sqrt(sample_size)

# Calculate the z-score for the given confidence level
z_score = norm.ppf((1 + confidence_level) / 2)
print(z_score)

# Calculate the margin of error
margin_of_error = z_score * standard_error

# Calculate the confidence interval
lower_limit = sample_mean - margin_of_error
upper_limit = sample_mean + margin_of_error

print(f"90% Confidence Interval: ({lower_limit:.4f}, {upper_limit:.4f})")

# Plotting the normal distribution
x = np.linspace(sample_mean - 4 * sample_std_dev, sample_mean + 4 * sample_std_dev, 1000)
plt.plot(x, norm.pdf(x, sample_mean, sample_std_dev), label='Normal Distribution')
plt.fill_between(x, norm.pdf(x, sample_mean, sample_std_dev), where=((x >= lower_limit) & (x <= upper_limit)), color='orange', alpha=0.5, label='90% Confidence Interval')
plt.xlabel('Breaking Load (units)')
plt.ylabel('Probability Density')
plt.title('Normal Distribution and 90% Confidence Interval')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Given data
sample_mean = -75.4
sample_std_dev = 10.9
sample_size = 25
confidence_level = 0.90

# Calculate the standard error of the mean
standard_error = sample_std_dev / np.sqrt(sample_size)

# Calculate the degrees of freedom (n-1)
degrees_of_freedom = sample_size - 1

# Calculate the t-score for the given confidence level and degrees of freedom
t_score = t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)

# Calculate the margin of error
margin_of_error = t_score * standard_error

# Calculate the confidence interval
lower_limit = sample_mean - margin_of_error
upper_limit = sample_mean + margin_of_error

print(f"90% Confidence Interval: ({lower_limit:.4f}, {upper_limit:.4f})")

# Plotting the t-distribution
x = np.linspace(sample_mean - 4 * sample_std_dev, sample_mean + 4 * sample_std_dev, 1000)
plt.plot(x, t.pdf(x, degrees_of_freedom, sample_mean, standard_error), label='t-Distribution')
plt.fill_between(x, t.pdf(x, degrees_of_freedom, sample_mean, standard_error), where=((x >= lower_limit) & (x <= upper_limit)), color='orange', alpha=0.5, label='90% Confidence Interval')
plt.xlabel('Breaking Load (units)')
plt.ylabel('Probability Density')
plt.title('t-Distribution and 90% Confidence Interval')
plt.legend()
plt.grid(True)
plt.show()
```
